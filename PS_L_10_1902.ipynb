```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data for height and weight
height = np.array([150, 155, 160, 165, 170, 175, 180, 185, 190]).reshape(-1, 1)
weight = np.array([50, 52, 56, 58, 62, 66, 70, 74, 78])

# Fit the linear regression model
model = LinearRegression()
model.fit(height, weight)

# Predict the weight for 172 cm
predicted_weight = model.predict(np.array([[172]]))
print("Predicted weight for 172 cm:", predicted_weight[0])

# Plot the data points and regression line
plt.scatter(height, weight, color="blue", label="Data Points")
plt.plot(height, model.predict(height), color="red", label="Regression Line")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
plt.title("Height vs Weight - Linear Regression")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data for horsepower, mileage, age, and price
X = np.array([
    [120, 35, 5],
    [150, 30, 4],
    [130, 33, 3],
    [160, 28, 2],
    [110, 38, 6],
    [140, 32, 4]
])
y = np.array([20, 25, 27, 30, 18, 24])

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict the price for given car features
features = np.array([[135, 31, 3]])
predicted_price = model.predict(features)
print("Predicted price for 135 HP, 31 mpg, 3 years old:", predicted_price[0])

# Importance of each feature (coefficients)
print("Feature importance (coefficients):", model.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Data for sunlight and plant height
sunlight = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
height = np.array([5, 9, 15, 23, 30, 32, 31, 29])

# Fit the polynomial regression model (degree 2)
degree = 2
poly_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
poly_model.fit(sunlight, height)

# Predict plant height for 6.5 hours of sunlight
predicted_height = poly_model.predict(np.array([[6.5]]))
print("Predicted plant height for 6.5 hours of sunlight:", predicted_height[0])

# Plot the original data and fitted curve
plt.scatter(sunlight, height, color="blue", label="Data Points")
plt.plot(sunlight, poly_model.predict(sunlight), color="red", label="Fitted Curve")
plt.xlabel("Sunlight (hours/day)")
plt.ylabel("Plant Height (cm)")
plt.legend()
plt.title("Sunlight vs Plant Height - Polynomial Regression")
plt.show()
```
