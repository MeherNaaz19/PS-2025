**PROBABILITY & STATISTICS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
file_path = "/content/employee_data_250-2.csv"
df = pd.read_csv(file_path)
```

```python
sns.set_style("whitegrid")
```

# 1. Histogram Analysis: Salary Distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Salary"], bins=10, kde=True, color="blue", edgecolor="black")
plt.xlabel("Salary (USD)")
plt.ylabel("Frequency")
plt.title("Salary Distribution")
plt.show()
```

Identify the most frequent salary range

```python
salary_counts, bin_edges = np.histogram(df["Salary"], bins=10)
most_frequent_range = (bin_edges[salary_counts.argmax()], bin_edges[salary_counts.argmax() + 1])
print(f"Most common salary range: {most_frequent_range}")
```

 # 2. Scatterplot: Experience vs. Salary

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Experience (Years)"], y=df["Salary"], color="blue", alpha=0.6)
sns.regplot(x=df["Experience (Years)"], y=df["Salary"], scatter=False, color="red")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary (USD)")
plt.title("Experience vs. Salary")
plt.show()
```

Calculate correlation between Experience and Salary

```python
correlation = df["Experience (Years)"].corr(df["Salary"])
print(f"Correlation between Experience and Salary: {correlation}")
```

# 3. Bar Chart: Employee Count by Department

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df["Department"], palette="viridis", order=df["Department"].value_counts().index)
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Employee Count by Department")
plt.xticks(rotation=45)
plt.show()
```

Find the department with the highest number of employees

```python
highest_department = df["Department"].value_counts().idxmax()
print(f"Department with the highest number of employees: {highest_department}")
```

# 4. Grouped Bar Chart: Job Satisfaction across Departments

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df["Department"], hue=df["Job Satisfaction"], palette="coolwarm")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Job Satisfaction Levels Across Departments")
plt.legend(title="Job Satisfaction")
plt.xticks(rotation=45)
plt.show()
```
